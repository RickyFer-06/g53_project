```python
import pandas as pd
import sqlite3
import os

df = pd.read_csv('G53_data (1).csv')

if not os.path.exists('data'):
    os.makedirs('data')

conn = sqlite3.connect('data/business.db')

corp = df[['corporation_id', 'corporation_name', 'corporation_comments']].drop_duplicates()
corp.columns = ['id', 'name', 'comments']
corp.to_sql('Corporation', conn, if_exists='replace', index=False)

broker = df[['broker_id', 'broker_name', 'corporation_id']].drop_duplicates()
broker.columns = ['id', 'name', 'corporation_id']
broker['license_number'] = 1000 + broker['id'] 
broker = broker[['id', 'name', 'license_number', 'corporation_id']]
broker.to_sql('Broker', conn, if_exists='replace', index=False)

client = df[['client_id', 'client_name', 'client_address']].drop_duplicates()
client.columns = ['id', 'name', 'address']
client.to_sql('Client', conn, if_exists='replace', index=False)


trade = df[['transaction_id', 'broker_id', 'client_id', 'transaction_date', 'investment_amount']].copy()
trade.columns = ['id', 'broker_id', 'client_id', 'date', 'amount']
trade['name'] = "Investment" 
trade = trade[['id', 'broker_id', 'client_id', 'name', 'date', 'amount']]
trade.to_sql('Trade', conn, if_exists='replace', index=False)

conn.close()
print("Sucesso! Base de dados business.db gerada com 4 tabelas: Corporation, Broker, Client e Trade.")
```
